```python
import pandas as pd

url = "https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/BankChurners.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Count of missing values
df.isnull().sum()
```

```python
# For numeric columns
df.describe()

# For categorical columns
df.describe(include='object')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='Attrition_Flag', data=df)
plt.title("Customer Churn vs Existing")
plt.show()
```

```python
sns.histplot(df['Customer_Age'], kde=True)
plt.title("Customer Age Distribution")
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python

numeric_cols = ['Customer_Age', 'Credit_Limit', 'Total_Revolving_Bal', 'Total_Trans_Amt', 'Total_Trans_Ct']

mean_values = df[numeric_cols].mean()
print("ðŸ“Š Mean:\n", mean_values)

median_values = df[numeric_cols].median()
print("\nðŸ“Š Median:\n", median_values)

variance_values = df[numeric_cols].var()
print("\nðŸ“Š Variance:\n", variance_values)
```

```python


1. Dataset Overview:
   - Total Records: 10,127
   - Target Variable: `Attrition_Flag` (Existing or Attrited Customers)
   - No missing values found.

2. Descriptive Statistics:
   - Average Customer Age: ~46 years
   - Average Credit Limit: ~8,600


3. Insights from Visualizations:
   - Majority of customers are **Existing**, with fewer Attrited (churned).
   - Age and gender distributions are balanced.
   - Customers with higher transaction activity tend to stay.
   - Strong correlation observed between `Total_Trans_Ct` and `Total_Trans_Amt`.

4. Churn Behavior:
   - Churn tends to occur among customers with lower credit limits or fewer transactions.

This EDA provides foundational understanding for modeling churn prediction in the future.
```
